```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x=np.linspace(0,5,11)
y=x**2
```

```python
plt.plot(x,y,'r')
plt.xlabel('x axis title here')
plt.ylabel('y axis title here')
plt.title('String title here')
plt.show()
```

```python
fig = plt.figure()
ax= fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('title')
ax.set_xlabel('x')
ax.set_ylabel('y')
```
